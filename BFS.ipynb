**Breadth Frst Search(BFS):**

BFS is a traversing algorithm where one should start traversing from a selected node (source or starting node) and traverse the graph layerwise thus exploring the neighbour nodes (nodes which are directly connected to source node). You must then move towards the next-level neighbour nodes.<br>

Queue is used here.<br>



---



*   *Time Complexity :* $O(|V|+|E|)$, where $|V|$ the number of vertices and $|E|$ is the number of edges in the graph. Note that $O(|E|)$} may vary between $O(1)$ (Adjacency List is used) and $O(|V|^2)$ (When adjacencyy matrix is used)
*   *Space COmplexity:* $O(|V|)$, where $V$ is the number of Vertices.

---





*Algorithm*<br>
* Step 1: 
Start from the starting node, mark it visited, explore its
adjacent nodes, put its adjacent in the queue.
* Step 2: 
Point one from the adjacent as the currently working node. The current node should be marked visited and poped while its adjacents are being explored and being queued. If there is no adjacent, then pop that node.
* Step 3: Repeat Steps 4 and 5 until
QUEUE is empty.
* Step 6: EXIT

```python
import collections as c

def BFS(graph, startNode):
  visited = set()              # initiate empty set
  queue = c.deque([startNode]) # initiate deque,
                               # its speciality is that, it can be popped/appended from both sides
  visited.add(startNode)       # put the starting node in the visited set
  print("BFS : ")
  while queue:  #not empty
    currentNode = queue.popleft()       #pop the current node from queue
    print(str(currentNode) + " ")

    for adjacent in graph[currentNode]:     # for an adjacent node in that current exploring node
      if adjacent not in visited:           # if an adjacent node is not visited
        visited.add(adjacent)               # mark that node as visited
        queue.append(adjacent)              # append that node to explore later

if __name__ == '__main__':
    graph1 = {0: [1, 2], 1: [2], 2: [3], 3: [1, 2]}
    graph2 = {'A': ['B', 'C'], 'B': ['D', 'E'], 'C': ['F'], 'D': [], 'E':['F'], 'F': [] }
    print("Following is Breadth First Traversal: ")
    BFS(graph1, 0)
    BFS(graph2, 'A')


```
